```python
#load up the basics
import os, shutil
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#read in the master DF
mdf = pd.read_csv('../data/tables/MasterDF_09102021.csv')

mdf.head()
```

```python
#read in new storm  catalog
newStorm = pd.read_csv('/data/NOAA_ERI/catalogs/Henri.csv')
newStorm.head()
```

```python
#update the new storm df to make it compliant

#add storm to the filename

newStorm['Filename'] = 'Henri/' + newStorm['file']

#remove col.
del newStorm['file']
del newStorm['Unnamed: 0']
del newStorm['geom_checksum']

newStorm.head()
```

```python
combined = pd.concat([mdf, newStorm])
```

```python
combined.head()
```

```python
#new csv to save
newmdfcsv = '../data/tables/MasterDF_09102021a.csv'

combined.to_csv(newmdfcsv, index=False)
```

```python
preds = pd.read_csv(newpredcsv)
preds.head()
```

```python
# make a matrix based on model 1 and model 2

preds['Model_One_Wash'] = np.where(preds['prediction3'] > 0.5, True, False)
preds['Model_Two_Wash'] = np.where(preds['prediction4'] > 0.5, True, False)
preds.head()
```

```python
#make a confuson matrix w/ crosstab
#For florence, 2062 are both false, 1208 are both true
#153 went from true to false
#218 went from false to true


confusion_matrix = pd.crosstab(preds['Model_One_Wash'], preds['Model_Two_Wash'], rownames=['Model 1'], colnames=['Model 2'])
print (confusion_matrix)
```

```python
#add the path
preds['Image_Path'] = aerial_folder + preds['image'].astype(str)
preds['GC1_Path'] = GC1path + preds['image'].astype(str)
preds['GC2_Path'] = GC2path + preds['image'].astype(str)
preds['Y_Path'] = gradcam_image_pathY + preds['image'].astype(str) + '.pdf'
preds['N_Path'] = gradcam_image_pathN + preds['image'].astype(str) +'.pdf'
#preds.head()
```

```python
#select New Yes (false then true)
New_Yes_preds = preds[(preds.Model_One_Wash == False) & (preds.Model_Two_Wash == True)]

#New_Yes_preds

#select New No (true then false)
New_No_preds = preds[(preds.Model_One_Wash == True) & (preds.Model_Two_Wash == False)]
```

```python
#do the new yes

#move those on the list
for index, row in New_Yes_preds.iterrows():
    #make figure
    plt.figure(figsize=(20,20))
    plt.subplot(131)
    plt.imshow(mpimg.imread(row['Image_Path']))
    plt.title('Original Image')
    plt.subplot(132)
    plt.imshow(mpimg.imread(row['GC1_Path']))
    plt.title('Model 2: {}'.format(row['prediction3']))
    plt.subplot(133)
    plt.imshow(mpimg.imread(row['GC2_Path']))
    plt.title('Model 3: {}'.format(row['prediction4']))

    #save figure
    plt.savefig(row['Y_Path'], format='pdf')
    plt.close()
```
